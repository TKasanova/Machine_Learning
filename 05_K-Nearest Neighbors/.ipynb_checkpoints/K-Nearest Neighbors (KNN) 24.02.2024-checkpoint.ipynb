```python
def manhattan(x, y):
    """ Manhattan Distance """
    d = np.sum(np.abs(x - y))
    return d
```

```python
x = np.array([3, 5])
print("x:", x)

y = np.array([6, 9])
print("y:", y)

d_man = manhattan(x, y)
print(d_man)
```

```python
def euclidean(x, y):
    """ Euclidean distanc """
    d = np.sqrt(np.sum(np.square(x - y)))
    return d
```

$$d_{euc}= \sqrt{\sum_{i=1}^{N}(X_{i}-Y_{i})^{2}}$$

```python
x = np.array([3, 5])
print("x:", x)

y = np.array([6, 9])
print("y:", y)

d_euc = euclidean(x, y)
print(d_euc)
```

## Elbow Method for Choosing Reasonable K Values

```python
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score
```

```python
test_error_rates = []
train_error_rates = []

for k in range(1,30):
    knn_model = KNeighborsClassifier(n_neighbors=k)
    knn_model.fit(X_train_scaled,y_train) 
   
    y_pred_test = knn_model.predict(X_test_scaled)
    y_pred_train = knn_model.predict(X_train_scaled)
    
    test_error = 1 - accuracy_score(y_test,y_pred_test)
    test_error_rates.append(test_error)
    
    train_error = 1 - accuracy_score(y_train,y_pred_train)
    train_error_rates.append(train_error)
```

```python
plt.figure(figsize=(15,8))
plt.plot(range(1,30), test_error_rates, color='blue', linestyle='--', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K_values')
plt.ylabel('Error Rate')

plt.xticks(range(0, 31, 1));
```

```python
plt.figure(figsize=(15,8))
plt.plot(range(1,30), test_error_rates, color='blue', linestyle='--', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K_values')
plt.ylabel('Error Rate')

plt.hlines(y=0.073, xmin = 0, xmax = 30, colors= 'r', linestyles="--")
plt.hlines(y=0.0505, xmin = 0, xmax = 30, colors= 'r', linestyles="--")
plt.hlines(y=0.057, xmin = 0, xmax = 30, colors= 'r', linestyles="--")

plt.xticks(range(0, 31, 1));
```

```python
plt.figure(figsize=(15,8))
plt.plot(range(1,30), test_error_rates, color='blue', linestyle='--', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K_values')
plt.ylabel('Error Rate')

plt.plot(range(1,30), train_error_rates, color='green', linestyle='--', marker='o',
         markerfacecolor='blue', markersize=10)

plt.xticks(range(0, 31, 1));
```

## Scores by Various K Values

```python
def eval_metric(model, X_train, y_train, X_test, y_test):
    y_train_pred = model.predict(X_train)
    y_pred = model.predict(X_test)
    
    print("Test_Set\n")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    print()
    print("Train_Set\n")
    print(confusion_matrix(y_train, y_train_pred))
    print(classification_report(y_train, y_train_pred))
```

```python
knn = KNeighborsClassifier(n_neighbors=1)

knn.fit(X_train_scaled, y_train)
print("WITH K=1\n")

eval_metric(knn, X_train_scaled, y_train, X_test_scaled, y_test)
```

```python
k_list = [3, 6, 9, 16, 21, 22, 23]
for i in k_list:
    knn = KNeighborsClassifier(n_neighbors= i)
    knn.fit(X_train_scaled,y_train)
    print(f'WITH K={i}\n')
    eval_metric(knn, X_train_scaled, y_train, X_test_scaled, y_test)
    print("______________________________________________________")
```

## Pipe Line oluturma

from sklearn.pipeline import Pipeline

from sklearn.impute import SimpleImputer

from sklearn.preprocessing import StandardScaler, OneHotEncoder

from sklearn.decomposition import PCA

from sklearn.feature_selection import SelectKBest, chi2

from sklearn.ensemble import RandomForestClassifier

### Pipeline iin admlar listesi
steps = [
 *    ('imputer', SimpleImputer(strategy='mean')), # Eksik deerleri doldurma
 *    ('scaler', StandardScaler()), # zellik leklendirme
 *    ('encoder', OneHotEncoder()), # Kategorik deikenleri kodlama
 *    ('pca', PCA(n_components=2)), # Boyut indirgeme
 *    ('feature_selection', SelectKBest(chi2, k=10)), # En iyi 10 zellii seme
 *    ('classifier', RandomForestClassifier()) # Snflandrma modeli
]

### Pipeline instance oluturma
pipeline = Pipeline(steps=steps)

## Neden Data Leakage olur?

apraz dorulama, veri setini bir dizi eitim ve test setine ayrr. Eer leklendirme ilemi tm veri seti zerinde apraz dorulamadan nce yaplrsa, test seti bilgileri (rnein, zelliklerin ortalamas ve standart sapmas) leklendirme ilemi srasnda dolayl olarak eitim setine "szdrlr". Bu, modelin test seti zerindeki gerek performansn doru bir ekilde lmeyi zorlatrr, nk model, aslnda test seti hakknda nceden bilgi sahibi olur.

## Doru yaklam nasl olmal?

Doru yaklam, leklendirme ilemini apraz dorulamann her bir iterasyonunda, yalnzca o iterasyonda kullanlan eitim verisi zerinde yapmaktr. Bylece, her bir test seti, model eitimi srasnda hi grlmemi olarak kalr ve data leakage nlenmi olur. Bu yaklam uygulamann en kolay yolu, leklendiriciyi ve modeli bir Pipeline nesnesi iinde birletirmek ve bu pipeline' apraz dorulama iin kullanmaktr.
